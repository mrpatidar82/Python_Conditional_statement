```python
#13. Find the factorial of a number using a for loop

fact=1
for i in range(1,6):
    fact=fact*i
    print(fact)    
```

```python
#14. Reverse a given number using a for loop.  Example: 1234 → 4321
    

num = input("Enter number: ")
rev = " " #empty string lega taki number ka addition na kar de esliye
for i in num:
    rev = i + rev

print("Reverse:", rev)
```

```python
#15. Print each character of a string using a for loop.---Example: "hello"

a="Hello"
for ch in a:
    print(ch,end=" ")
```

```python
#16. Count vowels in a given string.
a=input("enter a string:")
count=0
for i in a:
    if i in "aeiouAEIOU":
        print("Vowels is:",i)
        count=count+1
print("The vowels in string is ",count)        
```

```python
#18. Print the square of numbers from 1 to 10.
for i in range(1,11):
    print(i,"square is:",i*i)
    
```

```python
#19. Print the cube of numbers from 1 to 10.
for i in range(1,11):
    print(i,"cube is:",i*i*i)
```

```python
'''20. Find the sum of digits of a number.
    Example: 123 → 6'''

num=input("Enter a number:")
sum=0
for i in num:
    sum=sum+int(i)
    print("The result is ",sum)
print(sum)    
```
