```python
##isalnum
a="HELLO"
b="123"
c="hello"
d="Hello how are you"
e=" hello "

print(a.isalnum())# string me sirf letters ya numbers hona chahiye
print(a.isalpha())#sirf alphabets hone chahiye
print(a.isnumeric())#sirf numbers hona chahiye tab true dega 
print(a.capitalize())#first letter ko capital kar dega 
print(a.casefold())#sabko lower case  kar dega 
print(a.center(11,"*"))#center me string fill kar dega 
print(a.count("L"))#count karega ki (L) kitni bar aaya he string me 
print(a.endswith("O"))#string ke last me (O) he to tru dega 
print(a.find("L"))#jis index pe (L) he vo index de dega agr vo 3or 4 pe he to sirf pahle vala index dega only 3
print(a.index("L"))#agar value na mile to error dega 
print(a.isdecimal())#agar decimal number hoge to true dega varna false dega 
print(b.isdigit())#agr digit honge to true dega varna false
print(a.islower())#check krega ki string lower case me he ya nhi tab true false dega
print(a.isupper())#check krega ki string upper case  me he ya nhi tb true false dega 
print(a.lower())#string agr upper case me hogi to use vah lower case me bana dega 
print(a.lstrip())# left side space ko hata dega jaise ki yaha left side space ko remove kar diya esne 
print(c.replace("l","r"))#replace kar dega l ki jagah r kar dega 
print(c.rfind("l"))#last l kis index pe aa rha he vo btayega sirf
print(c.rindex("l"))#last index no. bta raha he ye l ka
print(c.split())
print(d.split())#string ko tod ke list bana deta he 
print(c.startswith("h"))#check karta he string kis se start ho rhi he 
print(e.strip())#left +right dono side ki space  hata dega
print(c.upper())#sab letters ko capital kar dega 


```
