```python
#1.Print all elements of a tuple using a for loop Input: (1, 2, 3, 4) Output: 1 2 3 4
a = (1, 2, 3, 4)
for i in a:
    print(i, end=" ")
```

```python
#2.Print all elements using while loop Input: (5, 6, 7) Output: 5 6 7
a = (5, 6, 7)
i = 0
while i < len(a):
    print(a[i], end=" ")
    i += 1
```

```python
#3.Find length of tuple (without using len) Input: (10, 20, 30) Output: 3
t = (10, 20, 30)
count = 0
for i in t:
    count += 1
print(count)
```

```python
#4.Find sum of all elements Input: (1, 2, 3, 4) Output: 10
t = (1, 2, 3, 4)
print(sum(t))
```

```python
#5.Find maximum element Input: (3, 7, 2, 9, 5) Output: 9
t = (3, 7, 2, 9, 5)
print(max(t))
```

```python
#6.Find minimum element Input: (3, 7, 2, 9, 5) Output: 2
t = (3, 7, 2, 9, 5)
print(min(t))
```

```python
#7.Count occurrence of a number Input: Tuple = (1, 2, 2, 3, 2), Number = 2 Output: 3
t = (1, 2, 2, 3, 2)
print(t.count(2))
```

```python
#Create a new tuple by adding an element Input: (1, 2, 3), Add = 4 Output: (1, 2, 3, 4)
t = (1, 2, 3)
t = t + (4,)
print(t)
```

```python
#9.Insert an element at specific index Input: Tuple = (1, 2, 4), Element = 3, Index = 2 Output: (1, 2, 3, 4)
t = (1, 2, 4)
l = list(t)
l.insert(2, 3)
t = tuple(l)
print(t)
```

```python
#10.Remove first occurrence of an element Input: (1, 2, 3, 2), Remove = 2 Output: (1, 3, 2)
t = (1, 2, 3, 2)
l = list(t)
l.remove(2)
t = tuple(l)
print(t)
```

```python
#11.Remove last element Input: (10, 20, 30, 40) Output: (10, 20, 30)
t=(10,20,30,40)
t=t[:-1]
print(t)
```

```python
t=(10,20,30,40)
l=list(t)
l.pop()
t=tuple(l)
print(t)
```

```python
#12.Reverse a tuple Input: (1, 2, 3) Output: (3, 2, 1)
t=(1,2,3)
print(t[::-1])
```

```python
#13.Copy a tuple Input: (1, 2, 3) Output: (1, 2, 3)
t=(1,2,3)
t2=tuple(t)
print(t2)
```

```python
#14.Merge two tuples Input: (1, 2), (3, 4) Output: (1, 2, 3, 4)
t=(1,2)
t2=(3,4)
print(t+t2)
```

```python
#15.Remove duplicates from tuple Input: (1, 2, 2, 3, 1) Output: (1, 2, 3)
t=(1,2,2,3,1)
res=[]
for i in t:
    if i not in res:
        res.append(i)
t=tuple(res)
print(t)
```

```python
#16.Sort tuple in ascending order Input: (5, 2, 8, 1) Output: (1, 2, 5, 8)
a=(5,2,8,1)
a=tuple(sorted(a))
print(a)
```

```python
#17.Sort tuple in descending order Input: (5, 2, 8, 1) Output: (8, 5, 2, 1)
a=(5,2,8,1)
a=tuple(sorted(a,reverse=True))
print(a)
```

```python
#18.Find second largest element Input: (10, 20, 5, 30) Output: 20
a=(10,20,5,30)
a=sorted(a)
print(a[-2])
```

```python
#20.Count even numbers Input: (1, 2, 3, 4, 5, 6) Output: 3
t=(1,2,3,4,5,6)
count=0
for i in t:
    if i%2==0:
        count+=1
print(count)
```

```python
t=(1,2,3,4,5,6)
output=sum(1 for i in t if i%2==0)
print(output)
```

```python
#21.Count odd numbers Input: (1, 2, 3, 4, 5) Output: 3
t=(1,2,3,4,5,6)
count=0
for i in t:
    if i%2!=0:
        count+=1
print(count)
```

```python
t=(1,2,3,4,5,6)
output=sum(1 for i in t if i%2!=0)
print(output)
```

```python
#22.Create new tuple with squares of elements Input: (1, 2, 3) Output: (1, 4, 9)
t=(1,2,3)
t2=tuple(i*i for i in t)
print(t2)
```

```python
t=(1,2,3)
res=[]

for i in t:
    square =i*i
    res.append(square)
t=tuple(res)
print(t)
```

```python
#23.Remove all negative numbers Input: (1, -2, 3, -4, 5) Output: (1, 3, 5)
t = (1, -2, 3, -4, 5)
t2 = tuple(i for i in t if i >= 0)
print(t2)
```

```python
t = (1, -2, 3, -4, 5)
res=[]
for i in t:
    if i>=0:
        res.append(i)
print(res)
t=tuple(res)
print(t)
```

```python
#24.Find index of an element (without using index()) Input: Tuple = (10, 20, 30), Element = 20 Output: 1
t = (10, 20, 30)
for i in range(len(t)):
    if t[i] == 20:
        print(i)
```

```python
#25.Rotate tuple right by 1 position Input: (1, 2, 3, 4) Output: (4, 1, 2, 3)
t = (1, 2, 3, 4)
t = (t[-1],) + t[:-1]
print(t)
```

```python
#26.Find common elements between two tuples Input: (1, 2, 3), (2, 3, 4) Output: (2, 3)
t1 = (1, 2, 3)
t2 = (2, 3, 4)
t = tuple(i for i in t1 if i in t2)
print(t)
```

```python
t1 = (1, 2, 3)
t2 = (2, 3, 4)
t3=()
for i in t1:
    if i in t2:
        t3=t3+(i,)
print(t3)
```

```python
#27.Flatten a nested tuple (1 level) Input: (1, (2, 3), 4) Output: (1, 2, 3, 4)
t = (1, (2, 3), 4)
res = []
for i in t:
    if isinstance(i, tuple):
        res.extend(i)
    else:
        res.append(i)
print(tuple(res))
```
