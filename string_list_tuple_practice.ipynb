```python
#39 Group Words by Length Write a program to group words based on their lengths. Input: ["hi","hello","hey"] Output: {2:['hi'], 3:['hey'], 5:['hello']}
lst = ["hi","hello","hey"]
res = {}
for i in lst:
    l = len(i)
    if l not in res:
        res[l] = []
    res[l].append(i)
print(res)
```

```python
#40 Longest Common Prefix Write a program to find the longest common prefix among a list of strings. Input: ["flower","flow","flight"] Output: "fl"
a = ["flower","flow","flight"]
b = a[0]
for i in a[1:]:
    while not i.startswith(b):
        b = b[:-1]
print(b)
```
