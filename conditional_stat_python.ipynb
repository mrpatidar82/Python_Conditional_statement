# Definition:
Conditional statements are statements that allow a program to make decisions based on a condition and execute specific code accordingly.

-> Simple Explanation

 If the condition is True → the code runs
 If the condition is False → the code is skipped (or another block runs)



Here, the condition age >= 18 is True, so the statement executes.

-> Types of Conditional Statements
if → checks a condition
if-else → two choices
if-elif-else → multiple conditions

```python
#1. Take two numbers from the user and print their sum, difference, product, division, and floor division.
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

print("sum is:", a + b)
print("difference is:", a - b)
print("product is:", a * b)
print("division is:", a / b)
print("floor division is:", a // b)
```

```python
'''2. Ask the user to enter a number x. Use the sep optional argument to print out x, 2x, 3x, 4x, and 5x, each separated by three dashes, like below.
Enter a number: 7
7 --- 14 --- 21 --- 28 --- 35'''
x = int(input("Enter a number: "))
print(x, 2*x, 3*x, 4*x, 5*x, sep=" --- ")
```

```python
'''#3. Write a program that asks the user for a weight in kilograms and
converts it to pounds. There are 2.2 pounds in a kilogram.'''

kg = float(input("Enter weight in kg: "))
pounds = kg * 2.2
print("weight in pounds:", pounds)
```

```python
'''#4. The distance between two cities (in km.) is input through the keyboard.
    Write a program to convert and print this distance in meters, feet, inches and centimeters.'''

km = float(input("Enter distance in km: "))
m = km * 1000
cm = km * 100000
feet = km * 3280.84
inch = km * 39370.1
print("distance in meter is:",m)
print("distance in feet is:",cm)
print("distance in inches is:",feet)
print("distance in centimeters is:",inch)
```

```python
#5. Demonstrate difference between == and is. 
a=[1,2,3]
b=[1,2,3]
print("The result is:",a==b)
print("The result is:",a is b)
```

```python
#21. Check if a number is positive, negative, or zero.
num=int(input("Enter a number:"))
if num>0:
    print(num,"is positive")
elif num<0:
    print(num,"is negative")
elif num==0:    
    print(num,"is zero")
else:
    print("enter valid number")
```

```python
#22. Check if a number is divisible by 2, 3, or neither.
num=int(input("Enter a number:"))
if num%2==0:
    print(num,"is divisible by 2")
elif num%3==0:
    print(num,"is divisible by 3")
else:
    print(num,"not divisible by 2 or 3")
```

```python
'''23. Check the grade based on marks:

Marks ≥ 90 → A
Marks ≥ 75 → B
Else → C'''
marks=int(input("Enter a marks:"))
if marks>=90:
    print(marks,"A grade")
elif marks>=75:
    print(marks,"B grade")
else:
    print(marks,"C garde")
```

```python
'''#24. Check age category:

Age < 13 → Child
Age < 20 → Teen
Else → Adult'''

age=int(input("enter your age:"))
if age>0 and age<150:
    if age<13:
        print(age,"child")
    elif age<20:
        print(age,"Teen")
    else:
        print(age,"adult")
else:
    print(age,"please enter age between 0 or 100")
```
