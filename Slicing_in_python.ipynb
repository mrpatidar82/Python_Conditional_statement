```python
'''Check if two strings are equal (without using ==)
Input: "hello", "hello"
Output: Equal'''

a="hello"
b="hello"

if len(a)!=len(b):
    print("not equal")
else:
    result=True
    for i in  range(len(a)):
        if a[i] != b[i]:
            result=False
            break
    if result:
        print("Equal")
    else:
        print("Not equal")
```

```python
'''25.Find longest repeating character sequence
Input: "aaabbccccd"
Output: cccc'''


s = "aaabbccccd"

max_seq = ""
curr = s[0]

for i in range(1, len(s)):
    if s[i] == s[i-1]:
        curr += s[i]
    else:
        if len(curr) > len(max_seq):
            max_seq = curr
        curr = s[i]

if len(curr) > len(max_seq):
    max_seq = curr

print(max_seq)
```

```python
'''Shift each character by 1
Input: "abc"
Output: bcd'''

a = "abc"
res = ""
for i in a:
    res += chr(ord(i) + 1)
print(res)
```
