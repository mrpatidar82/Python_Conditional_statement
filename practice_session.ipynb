```python
#6 Reverse a number using a while loop.
num = int(input("Enter number: "))
rev = 0
while num > 0:
    digit = num % 10
    rev = rev * 10 + digit
    num = num // 10
print("Reversed number:", rev)
```

```python
#7. Check whether a number is palindrome using loop and condition.
num = int(input("Enter number: "))
real = num
rev = 0

while num > 0:
    digit = num % 10
    rev = rev * 10 + digit
    num = num // 10

if real == rev:
    print("Palindrome")
else:
    print("Not Palindrome")
```

```python
'''16. Create a menu-driven program using while loop:

1 → Add two numbers
2 → Subtract
3 → Exit'''

while True:
    choice = int(input("1.Add  2.Subtract  3.Exit: "))

    if choice == 1:
        a = int(input("Enter first number: "))
        b = int(input("Enter second number: "))
        print("Result:", a + b)
    elif choice == 2:
        a = int(input("Enter first number: "))
        b = int(input("Enter second number: "))
        print("Result:", a - b)
    elif choice == 3:
        break
    else:
        print("Invalid choice")
```

```python
'''18. Take a list and:
Count even, odd, positive, negative numbers using one loop'''

nums = [10, -5, 0, 8, -2, 7, 4]
even = 0
odd = 0
pos = 0
neg = 0
for i in nums:
    if i >= 0:
        pos += 1
    else:
        neg += 1
    if i % 2 == 0:
        even += 1
    else:
        odd += 1
print("Even:", even)
print("Odd:", odd)
print("Positive:", pos)
print("Negative:", neg)
```

```python
'''19. Print a pyramid pattern:

   *
  ***
 *****'''
n = 3
for i in range(1, n+1):
    for j in range(n - i):
        print(" ", end="")
    for j in range(2*i - 1):
        print("*", end="")
    print()
```

```python
'''22. Simulate ATM system:

Balance check
Deposit
Withdraw (check insufficient balance using if-else)'''

balance = 0
while True:
    print("1. Check Balance","2. Deposit","3. Withdraw","4. Exit")
    choice = int(input("Enter choice: "))

    if choice == 1:
        print("Balance:", balance)

    elif choice == 2:
        amt = int(input("Enter amount: "))
        balance += amt

    elif choice == 3:
        amt = int(input("Enter amount: "))
        if amt > balance:
            print("Insufficient Balance")
        else:
            balance -= amt
    elif choice == 4:
        break
    else:
        print("Invalid choice")
```

```python
#23. Check if a string is palindrome ignoring spaces and case.
string = input("Enter string: ")
a = ""
for i in string:
    if i != " ":
        a += i.lower()
rev = a[::-1]
if a == rev:
    print("Palindrome")
else:
    print("Not Palindrome")
```

```python
'''24. Create a number guessing game:
User keeps guessing until correct
Give hints: "Too High" / "Too Low"'''

secret = 7
while True:
    guess = int(input("Guess number: "))
    if guess > secret:
        print("Too High")
    elif guess < secret:
        print("Too Low")
    else:
        print("Correct Guess")
        break
```

```python
'''25. Process a list of student marks:

Print grade:

  * ≥90 → A
  * ≥75 → B
  * ≥50 → C
  * else → Fail
Also count how many students passed/failed'''

marks = [95, 80, 45, 60, 30]
pass_ = 0
fail_ = 0
for m in marks:
    if m >= 90:
        grade = "A"
        pass_ += 1
    elif m >= 75:
        grade = "B"
        pass_ += 1
    elif m >= 50:
        grade = "C"
        pass_ += 1
    else:
        grade = "Fail"
        fail_ += 1
    print(m, "-", grade)
print("Passed:", pass_)
print("Failed:", fail_)
```

```python
students = []
while True:
    name = input("Enter student name (or type 'exit' to stop): ")
    if name == "exit":
        break
    marks = int(input("Enter marks: "))
    
    if marks >= 90: # Grade assign
        grade = "A"
    elif marks >= 75:
        grade = "B"
    elif marks >= 50:
        grade = "C"
    else:
        grade = "Fail"

    # Store data
    students.append([name, marks, grade])

# Print all students
print("Student Records:")
for s in students:
    print("Name:", s[0], "| Marks:", s[1], "| Grade:", s[2])

# Find topper
top_marks = 0
topper = ""

for s in students:
    if s[1] > top_marks:
        top_marks = s[1]
        topper = s[0]

print("\nTopper:", topper, "-", top_marks)
```
