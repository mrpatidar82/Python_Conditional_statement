```python
#1.Print all elements of a list using a for loop Input: [1, 2, 3, 4] Output: 1 2 3 4
a= [1, 2, 3, 4]
for i in a:
    print(i,end=" ")
```

```python
#2.Print all elements using while loop Input: [5, 6, 7] Output: 5 6 7
a=[5,6,7]
i=0
while i<len(a):
    print(a[i],end=" ")
    i=i+1
```

```python
#4.Find maximum element in list Input: [3, 7, 2, 9, 5] Output: 9
a=[3,7,2,9,5]
res=a[0]
for i in a:
    if i>res:
        res=i
print(res)
```

```python
#6.Count occurrences of a number Input: List = [1, 2, 2, 3, 2], Number = 2 Output: 3
a = [1, 2, 2, 3, 2]
count = 0
for i in a:
    if i == 2:
        count += 1
print(count)
```

```python
#14.. Remove duplicates from listInput: [1, 2, 2, 3, 1] Output: [1, 2, 3]

a = [1, 2, 2, 3, 1]
b = []
for i in a:
    if i not in b:
        b.append(i)
print(b)
```

```python
#18.Check if element exists in list Input: List = [1, 2, 3], Element = 2 Output: Found
a = [1, 2, 3]
if 2 in a:
    print("Found")
```

```python
#19.Count even numbers Input: [1, 2, 3, 4, 5, 6] Output: 3
a = [1, 2, 3, 4, 5, 6]
count = 0
for i in a:
    if i % 2 == 0:
        count += 1
print(count)
```

```python
#20.Count odd numbers Input: [1, 2, 3, 4, 5] Output: 3
a = [1, 2, 3, 4, 5]
count = 0
for i in a:
    if i % 2 != 0:
        count += 1
print(count)
```

```python
#21.Create new list with squares of elementsInput: [1, 2, 3]Output: [1, 4, 9]
a = [1, 2, 3]
b = []
for i in a:
    b.append(i*i)
print(b)
```

```python
#22.Remove all negative numbersInput: [1, -2, 3, -4, 5]Output: [1, 3, 5]
a = [1, -2, 3, -4, 5]
b = []
for i in a:
    if i >= 0:
        b.append(i)
print(b)
```

```python
#23. Find index of element (without using index())Input: List = [10, 20, 30], Element = 20 Output: 1

a = [10, 20, 30]
for i in range(len(a)):
    if a[i] == 20:
        print(i)
```

```python
a = [10, 20, 30]
print(a.index(20))
```

```python
#24. Rotate list to right by 1 positionInput: [1, 2, 3, 4]Output: [4, 1, 2, 3]
a = [1, 2, 3, 4]
last = a[-1]
b = [last]
for i in range(len(a)-1):
    b.append(a[i])
print(b)
```

```python
a = [1, 2, 3, 4]
last = a.pop()
a.insert(0, last)
print(a)
```

```python
#25.Find common elements in two lists Input: [1, 2, 3], [2, 3, 4] Output: [2, 3]
a = [1, 2, 3]
b = [2, 3, 4]
c = []
for i in a:
    if i in b:
        c.append(i)
print(c)
```

```python
#26.
a = [1, [2, 3], 4]
b = []
for i in a:
    if type(i) == list:
        for j in i:
            b.append(j)
    else:
        b.append(i)
print(b)
```

```python
a = [1, [2, 3], 4]
b = []
for i in a:
    if isinstance(i, list):
        b.extend(i)
    else:
        b.append(i)
print(b)
```
